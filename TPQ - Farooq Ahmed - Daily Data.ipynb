## 2. Machine Learning Review  <a class="anchor" id="second-bullet"></a>

**2.1	Machine Learning** 
Machine learning (ML) is the scientific study of algorithms and statistical models that computer systems use to effectively perform a specific task without using explicit instructions, relying on patterns and inference instead. It is seen as a subset of artificial intelligence.Machine learning tasks are classified into several broad categories. In supervised learning, the algorithm builds a mathematical model from a set of data that contains both the inputs and the desired outputs. In unsupervised learning, the algorithm builds a mathematical model from a set of data which contains only inputs and no desired output labels. Supervised learning algorithms build a mathematical model of a set of data that contains both the inputs and the desired outputs. Classification algorithms and regression algorithms are types of supervised learning. Classification algorithms are used when the outputs are restricted to a limited set of values.Regression algorithms are named for their continuous outputs, meaning they may have any value within a range. [1](#1-ref)

**2.2	Machine Learning trading rules**
Data science and machine learning analysis and modelling methods are being increasingly employed in portfolio performance and portfolio risk modelling including trading. [2](#2-ref)
Machine learning is not a type of trading strategy. It is a tool kit for use to conduct analysis. However, it is popular enough to warrant its own section.[3](#3-ref) There are various Machine Learning models i.e. Naïve Bayes Classifier Algorithm,

**2.3 AdaBoost regressor**
An AdaBoost regressor is a meta-estimator that begins by fitting a regressor on the original dataset and then fits additional copies of the regressor on the same dataset but where the weights of instances are adjusted according to the error of the current prediction. As such, subsequent regressors focus more on difficult cases.[5](#5-ref)

**2.3 Bagging regressor**
A Bagging regressor is an ensemble meta-estimator that fits base regressors each on random subsets of the original dataset and then aggregate their individual predictions (either by voting or by averaging) to form a final prediction. Such a meta-estimator can typically be used as a way to reduce the variance of a black-box estimator (e.g., a decision tree), by introducing randomization into its construction procedure and then making an ensemble out of it.[6](#6-ref)

**2.4 Decision Trees (DTs)**
DTs are a non-parametric supervised learning method used for classification and regression. The goal is to create a model that predicts the value of a target variable by learning simple decision rules inferred from the data features.[7](#7-ref)

**2.5 Gradient Boosting for regression**
GB builds an additive model in a forward stage-wise fashion; it allows for the optimization of arbitrary differentiable loss functions. In each stage a regression tree is fit on the negative gradient of the given loss function.

**Random forest **
A random forest is a meta estimator that fits a number of classifying decision trees on various sub-samples of the dataset and uses averaging to improve the predictive accuracy and control over-fitting. The sub-sample size is always the same as the original input sample size but the samples are drawn with replacement if bootstrap=True.

#### Parameterization <a class="anchor" id="13-bullet"></a>

```python
inoutsample = '2017-06-30 22:00:00'

symbol = 'EUR=' #  input column name for symbol to be used.
laging = 10


seed = 7 # standard randon seed setting for all the models
leaf = 400 # min_samples_leaf setting for ML Models
estimator = 20 #n_estimators
cost = -0.00005 #  commission  cost percentage

# Reference to Pricing from Oanda 1 Lot is 5$ = or 5 /100000 = 0.00005 commission
# https://www1.oanda.com/register/docs/divisions/oc/price_sheet.pdf

```

Once, we created our new Pandas dataframe, we create log returns for the selected intrument. In this case its the returns series of EURUSD

```python
data['Returns'] = np.log(data / data.shift(1))
```

#### Features Modelling <a class="anchor" id="13a-bullet"></a> 
We then prepare the various input features which will be used by the machine learning to train itself for making the predictions. We define the following input features:
Lag Returns for 10 lags 
20 day moving average
60 day moving average
20 day maximum and minimum price
Standard deviation for a period of 5,10,20,60 days.

Additionally, we create two binary (0 / 1) signals based on SMAs and Returns. 



```python
lags = laging
cols = []
for lag in range(1, lags + 1):
    col = 'Lag_%d' % lag
    data[col] = data['Returns'].shift(lag)
    cols.append(col)
```

```python
data['SMA1'] = data['Close'].rolling(20).mean()
data['SMA2'] = data['Close'].rolling(60).mean()
data['Max'] = data['Close'].rolling(20).max().shift(1)
data['Min'] = data['Close'].rolling(20).min().shift(1)
data['Vol1'] = data['Close'].rolling(5).std().shift(1)
data['Vol2'] = data['Close'].rolling(10).std().shift(1)
data['Vol3'] = data['Close'].rolling(20).std().shift(1)
data['Vol4'] = data['Close'].rolling(60).std().shift(1)
data['SMA'] = np.where(data['SMA1'] > data['SMA2'], 1, 0)
data['SMA'] = data['SMA'].shift(1)

cols.append('Vol1')
cols.append('Vol2')
cols.append('Vol3')
cols.append('Vol4')
cols.append('Max')
cols.append('Min')
cols.append('SMA')
```

```python
for m in [5]: # , 10, 15]:
    col = 'MOM_%d' % m
    data[col] = np.where(data['Returns'].rolling(m).mean() > 0, 1, 0)
    data[col] = data[col].shift(1)
    cols.append(col)
```

Next, we drop all the rows storing NaN values by using the dropna() function.

```python
data.dropna(inplace=True)
```

###### Computing Strategy Returns  for In-sample Training Data<a class="anchor" id="17-bullet"></a> 
Now that we have the predicted values of the EURUSD movement. We can compute the returns of the strategy. We will be taking a long position when the predicted value of y is true and will take a short position when the predicted signal is False.
We first compute the returns that the strategy will earn if a long position is taken at the end of today, and squared off at the end of the next day open. 

Hence, to calculate each return whether the signal is profitable or not. We mutliple our position with returns. Note. the forecast position is 1 days lags, since not to have forward bias.

Similarlly, we add trascation cost after taking a positon as percentge of return. However, its a appoximate approach to understand the impact of cost. The cost basically minus from the gross return ( gross loss).

```python
train['OLSStrategy'] = (train['OLSPosition'] * train['Returns']) + cost 
```

```python
train['LMStrategy'] = (train['LMPosition'] * train['Returns']) + cost
```

```python
train['DTRStrategy'] = (train['DTRPosition'] * train['Returns']) + cost
```

```python
train['AddboostStrategy'] = (train['AddboostPosition'] * train['Returns']) + cost
```

```python
train['Bagging_regStrategy'] = (train['Bagging_regPosition'] * train['Returns']) + cost
```

```python
train['DTCStrategy'] = (train['DTCPosition'] * train['Returns']) + cost
```

```python
train['GradientboostStrategy'] = (train['GradientboostPosition'] .shift(1)  * train['Returns']) + cost
```

```python
train['RandomForestRegressorStrategy'] = (train['RandomForestRegressorPosition'] * train['Returns']) + cost
```

```python
train['RandomForestClassifierStrategy'] = (train['RandomForestClassifierPosition'].shift(1) * train['Returns']) + cost
# RandomForestClassifierStrategy required to be shift 1 day due to forward bias
```

###### Computing Strategy Returns   for Out-sample Test Data <a class="anchor" id="19c-bullet"></a> 
Now that we have the predicted values of the EURUSD movement. We can compute the returns of the strategy. We will be taking a long position when the predicted value of y is true and will take a short position when the predicted signal is False.
We first compute the returns that the strategy will earn if a long position is taken at the end of today, and squared off at the end of the next day open. 

Hence, to calculate each return whether the signal is profitable or not. We mutliple our position with returns. Note. the forecast position is 1 days lags, since not to have forward bias.

Similarlly, we add trascation cost after taking a positon as percentge of return. However, its a appoximate approach to understand the impact of cost. The cost basically minus from the gross return ( gross loss).

```python
test['OLStrategy'] = (test['OLSPosition'] * test['Returns']) + cost 
```

```python
test['LMStrategy'] = (test['LMPosition'] * test['Returns']) + cost
```

```python
test['DTRStrategy'] = (test['DTRPosition'] * test['Returns']) + cost
```

```python
test['AddboostStrategy'] = (test['AddboostPosition'] * test['Returns']) + cost
```

```python
test['Bagging_regStrategy'] = (test['Bagging_regPosition'] * test['Returns']) + cost
```

```python
test['DTCStrategy'] = (test['DTCPosition'] * test['Returns']) + cost
```

```python
test['GradientboostStrategy'] = (test['GradientboostPosition'].shift(1)   * test['Returns']) + cost
```

```python
test['RandomForestRegressorStrategy'] = (test['RandomForestRegressorPosition'] * test['Returns']) + cost
```

```python
test['RandomForestClassifierStrategy'] = (test['RandomForestClassifierPosition'].shift(1)   * test['Returns']) + cost
```
